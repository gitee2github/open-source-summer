```python
import jieba
import pandas as pd
import re
import math
import jieba.analyse
```

```python
words = []
with open('./File/THUOCL_it.txt', 'r', encoding='utf-8') as openFile:
  reads = openFile.readlines()
  for read in reads:
    key = read.split(' ')
    words.append(key[0])

with open('./File/test.txt', 'w', encoding='utf-8') as file:
  for item in words:
    file.write(f'{item} 10' + '\n')
```

```python
jieba.analyse.set_idf_path('./File/test.txt')
jieba.load_userdict('./File/THUOCL_it.txt')
```
