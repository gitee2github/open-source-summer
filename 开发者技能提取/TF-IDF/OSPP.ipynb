```python
import jieba
import pandas as pd
import re
import math
import jieba.analyse
```

```python
def filter_words(words, stopWords):
  filteredWord = []
  for word in words:
    if word in stopWords:
      continue
    elif word is '\n' or word is '' or word is ' ':
      continue
    else:
      filteredWord.append(word)

  return filteredWord
```

```python
stopWords = []
with open('./File/stop.txt', encoding='utf-8') as openFile:
  reads = openFile.readlines()
  for read in reads:
    stopWords.append(read.strip('\n'))
```

```python
jieba.re_han_default = re.compile("([\u4E00-\u9FD5a-zA-Z0-9+#&\._% ]+)", re.U)
```

```python
jieba.load_userdict('./File/userDict.txt')
```

```python
data = pd.read_csv('./File/OSPPEvent.txt')
dict = {}
```

```python
for item in data['event']:
     res = jieba.cut(str(item))
     resA = filter_words(res, stopWords)
     for temp in resA:
        if temp in dict:
            dict[temp] += 1
        else:
            dict[temp] = 1
     
```

```python
dictSort = {k: v for k, v in sorted(dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
eventCount = len(data['event'])
eventCount

dictWordInTxt = {}

for i in range(eventCount):
    print(i)
    txt = data['event'][i]
    for temp in dictSort.keys():
        res = jieba.cut(str(txt))
        resA = list(set(filter_words(res, stopWords)))
        if temp in resA:
            if temp in dictWordInTxt:
                dictWordInTxt[temp] += 1
            else:
                dictWordInTxt[temp] = 1
```

```python
# dictWordInTxt
with open('./File/dictWord.txt', 'w', encoding='utf-8') as file:
    for temp in dictWordInTxt.keys():
        file.write(f'{temp} {math.log(eventCount / (1.0 + dictWordInTxt[temp]))}' + '\n')
```

```python
jieba.analyse.set_idf_path('./File/dictWord.txt')
```
